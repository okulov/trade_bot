```python
import mplfinance as mpf
import pandas as pd
import talib
```

```python
daily = pd.read_csv('data/day_data.csv',index_col=0,parse_dates=True)[:100]
daily.index.name = 'Date'
daily.shape
daily.head(3)
daily.tail(3)
```

```python
macd = talib.MACD(daily['Close'])
```

```python
apds = [ 
    mpf.make_addplot(macd[0]),
    mpf.make_addplot(macd[1]),
    mpf.make_addplot(macd[2], panel=1),
]
```

```python
mpf.plot(daily,addplot=apds)
```
